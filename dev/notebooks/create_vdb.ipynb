```python
from ai.store import NLPFactory, VectorDBFactory
from ai.core import get_settings

SETTINGS = get_settings()
nlp_factory = NLPFactory(SETTINGS)
nlp = nlp_factory.create("gemini")
nlp.connect()

vectordb_factory = VectorDBFactory(SETTINGS)
vectordb = vectordb_factory.create("chroma")
vectordb.connect()
```

```python
import pandas as pd

df = pd.read_csv("data.csv")

list_text = list()
for _, row in df.iterrows():
    list_text.append(row["semantic_desc"])

metadata = df.to_dict(orient="records")
```

```python
embeddings = nlp.embed(list_text, 10, "gemini-embedding-001", "SEMANTIC_SIMILARITY")
```

```python
vectordb.create_collection("semantic_desc")
```

```python
from uuid import uuid4

ids = [str(uuid4()) for _ in range(len(list_text))]
```

```python
vectordb.add_points("semantic_desc",ids, embeddings, metadata)
```
